 ## 1. Optimally Weighted GMM Estimator

An optimally weighted GMM estimator is consistent with the smallest asymtotic variance of all GMM estimators. To construct any GMM estimator, we find a set of parameters that minimize the weighted sum of the square sample moments. All GMM estimators are consistent. 

To construct the optimal GMM estimator, we then select a weighting matrix that minimizes the asymptotic variance, making the GMM estimator the most efficient (among GMM estimators). To minimize the variance of all empirical moments, we weight each moment inversely (i.e. upweighting moments with low variance and downweighting moments with high variance).

#### Moment conditions and sample analogues:
$$
E[y - \mu] = 0;  \sum_{i=1}^{N}[y - \mu] = 0
$$
$$
E[y - \mu]^2 - \sigma^2 = 0; \sum_{i=1}^{N}[y - \mu]^2 - \sigma^2 = 0
$$
$$
E[y - \mu]^3 = 0; \sum_{i=1}^{N}[y - \mu]^3 = 0
$$

#### Optimally weighted GMM:
$$g_n(\mu, \sigma^2) = \sum_{i=1}^{N}[y - \mu], [y - \mu]^2 - \sigma^2,[y - \mu]^3 
$$
$$
J(\mu, \sigma^2) = [g_n(\mu, \sigma^2)]'W^{-1} [g_n(\mu, \sigma^2)]
$$
$$ 
W =\frac{\hat {e}'\hat {e}}{N}
$$
$$
(\mu, \sigma^2)_{GMM} = argmin_\mu, \sigma^2 J(\mu, \sigma^2)
$$

## 3. Data generating process for GMM estimators

```python
import numpy as np
import pandas as pd
import scipy.stats.distributions as iid
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
```

**a)**

```python
def dgp_a(mu, s2, N):
        y = iid.norm(loc=mu).rvs(size=(N,1))*s2**.5
        return y
```

```python
# test the function
mu_true = 5
s2_true = 1
N = 100
data_a = dgp_a(mu_true, s2_true, N)
```

**b)**

```python
def dgp_b(beta, N):
        # get everything into the right shape
        beta = np.array(beta).reshape(-1,1)
        k    = len(beta)
        
        # generate data
        X = iid.norm().rvs(size=(N, k))
        u = iid.norm().rvs(size=(N,1))

        y = X @ beta + u
        return y, X
```

```python
# test the function
beta_true = 5
N = 100
data_b = dgp_b(beta_true, N)
```

```python
data_df_b = pd.DataFrame(np.concatenate(data_b, axis=1), columns = ["y", "X"])
data_df_b.head()
```

```python
def dgp_d(beta, s, N):
        # get everything into the right shape
        beta = np.array(beta).reshape(-1,1)
        k    = len(beta)
        
        # generate data
        X = iid.norm().rvs(size=(N, k))
        u = iid.norm().rvs(size=(N,1))
        u = u*np.exp(X*s)**.5 

        y = X @ beta + u
        return y, X
```

```python
def dgp_e(beta, N):
    # get everything into the right shape
    beta = np.array(beta).reshape(1,-1)
    k    = len(beta)
      
    # generate data
    X = iid.norm().rvs(size=(N, k))
    
    u = iid.norm().rvs(size=(N,1))

    y = X @ beta.T + u
         
    Z = iid.norm().rvs(size=(N,1))
    Z = Z - X.T @ (X @ Z) @ beta / (X.T @ X @ beta) # something is wrong with the dimensions
    return y, X, Z
```

```python
def dgp_h(alpha, gamma, N):
        u = iid.norm().rvs(size=(N,1))
        y = (alpha + u)**(1/gamma)
        
        Z = iid.norm().rvs(size=(N,1))
        Z = Z - gamma@y**(gamma-1).T @ (Z @ X) 
        
        return y, Z
```

```python
# create moments
def gj(params, data):
    # inputs
    beta, s2 = params
    y, X = data
    # moments 
    moment1 = y - X*beta
    moment2 = (y - X*beta)**2 - s2
    # stack the moments next to each other
    moments = np.concatenate([moment1, moment2], axis=1)
    
    return moments
```

```python
# test out the function
gj([10, .01], data_3)[:5]
```

```python
def gN(params, data):
    e = gj(params, data)
    gN = e.mean(axis=0).reshape((len(params),1))
    return gN
```

```python
def invOmega(params, data):
    e = gj(params, data)
    # recenter
    e = e - e.mean(axis=0)
    N = e.shape[0]
    var = e.T@e/N
    return np.linalg.inv(var)
```

```python
invomega_1 = invOmega([beta_true, s2_true], data_1)
```

```python
invomega_2 = invOmega([beta_true, s2_true], data_2)
```

```python
invomega_3 = invOmega([beta_true, s2_true], data_3)
```

```python
# minimizer
def J(params,invomega,data):

    g = gN(params,data) 
    N = data[0].shape[0]

    return ((N*g.T)@invomega@g).squeeze() 
```
